```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Using seaborn to load the Iris Dataset
iris = sns.load_dataset('iris')
```

```python
# Print shape
print("Dataset Shape:", iris.shape)

# Print column names
print("\nColumn Names:", iris.columns.tolist())

# Display first 5 rows
print("\nFirst 5 Rows:\n", iris.head())

# Summary statistics
print("\nDataset Info:")
iris.info()

print("\nDescriptive Statistics:\n", iris.describe())
```

```python
# Visualize the Data
# Scatter Plot: Sepal Length vs. Sepal Width by Species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', style='species', data=iris)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.savefig('sepal_scatter_plot.png')
plt.show()
```

```python
# Histograms: Feature Distributions
iris.hist(figsize=(10, 8), bins=20)
plt.suptitle('Histograms of Iris Features')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('iris_histograms.png')
plt.show()
```

```python

# Box Plots: Identify Outliers by Species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Box Plot of Petal Width by Species')
plt.savefig('petal_width_boxplot.png')
plt.show()
```
